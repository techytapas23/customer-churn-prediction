# Lets Build the MachineLearning Model here:-

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
lr = LogisticRegression(random_state=42)

lr.fit(x_train_scaled, y_train)
```

```python
y_pred_lr = lr.predict(x_test_scaled)
```

```python
from sklearn.metrics import (
accuracy_score,
precision_score,
recall_score,
f1_score,
confusion_matrix,
classification_report
)
```

```python
accuracy_score(y_test,y_pred_lr)
```

```python
precision_score(y_test,y_pred_lr)
```

```python
recall_score(y_test,y_pred_lr)
```

```python
f1_score(y_test,y_pred_lr)
```

```python
#Lets calculate classification report
print(classification_report(y_test,y_pred_lr))
```

```python
#Lets calculate confusion matrix
cm = confusion_matrix(y_test,y_pred_lr)
cm
```

```python
#Lets Visualize the Confusion Matrix
plt.figure(figsize = (6,5))
sns.heatmap(
    cm,
    annot = True,
    fmt = 'd',
    cmap = 'Blues'
)

plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Logistic Regresion Confusion Matrix')
plt.show()
```

```python
# Lets Build the Decision Tree
dt = DecisionTreeClassifier(
    random_state = 42
)
```

```python
dt.fit(
    x_train,
    y_train
)
```

```python
y_pred_dt = dt.predict(x_test)
```

```python
print("Accuracy :",accuracy_score(y_test,y_pred_dt))
print(classification_report(y_test,y_pred_dt))
```

```python
# Confusion Matrix
cm = confusion_matrix(
    y_test,
    y_pred_dt
)

plt.figure(figsize = (6,5))
sns.heatmap(
    cm,
    annot = True,
    fmt = "d",
    cmap = "Greens"
)

plt.title("Decision Tree")
plt.show()
```

```python
# Now lets calculate on the basis of Random Forest:-
rf = RandomForestClassifier(
    n_estimators = 100, # Means 100 (Decision)trees randomly we will take to  train the model
    random_state = 42
)
```

```python
rf.fit(
    x_train,
    y_train
)
```

```python
y_pred_rf = rf.predict(x_test)
```

```python
print("Accuracy: ", accuracy_score(y_test, y_pred_rf))
print(classification_report(y_test, y_pred_rf))
```

```python
cn = confusion_matrix(
    y_test,
    y_pred_rf
)

plt.figure(figsize = (6,5))

sns.heatmap(
    cm,
    annot = True,
    fmt = 'd',
    cmap = "Oranges"
)

plt.title("Random Forest")
plt.show()
```

```python
# Now Lets compare all the models we have built
results = pd.DataFrame({
    "Model":[
        "Logistic Regression",
        "Decision Tree",
        "Random Forest"
    ],
    "Accuracy":[
        accuracy_score(y_test,y_pred_lr),
        accuracy_score(y_test,y_pred_dt),
        accuracy_score(y_test,y_pred_rf)
    ],
    "Precision":[
        precision_score(y_test,y_pred_lr),
        precision_score(y_test,y_pred_dt),
        precision_score(y_test,y_pred_rf)
    ],
    "Recall":[
        recall_score(y_test,y_pred_lr),
        recall_score(y_test,y_pred_dt),
        recall_score(y_test,y_pred_rf)
    ],
    "F1 Score":[
        f1_score(y_test,y_pred_lr),
        f1_score(y_test,y_pred_dt),
        f1_score(y_test,y_pred_rf)
    ]
})


results
```
